````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {p: int(input(f"Amount of {p}s: ")) for p in products}
    
    return inventory
````

````python
# 2. Modify the function get_customer_orders so it prompts the user to enter the number of customer orders and gathers the product names using a loop and user input. Use comprehension.


products = ["t-shirt", "mug", "hat", "book", "keychain"]

def get_customer_orders():
    num_orders = int(input("How many products do you wish to order? "))
    
    customer_orders = {
        order for _ in range(num_orders) 
        if (order := input("Enter product name: ")) in products 
        or print(f"Sorry, we don't have '{order}'!")
    }
            
    return customer_orders

# Test:
customer_orders = get_customer_orders() 
print("Your final order is:", customer_orders)
````

````python
# 3. Add a new function to calculate the total price of the customer order. For each product in customer_orders, prompt the user to enter the price of that product. Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.

def calculate_total_price(customer_orders):
    total_price = sum([float(input(f"Enter the price for {product}: ")) for product in customer_orders])
    
    print(f"The total price of your order is: ${total_price:.2f}")
    return total_price

# Test:
total = calculate_total_price(customer_orders)
````

````python
# 4. Modify the update_inventory function to remove the product from the inventory if its quantity becomes zero after fulfilling the customer orders. Use comprehension to filter out the products with a quantity of zero from the inventory.

def update_inventory(customer_orders, inventory):
    # Decreasing the quantity for each ordered product
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1
            print(f"Updated {product}: {inventory[product]} left in stock.")
        else:
            print(f"Warning: {product} is not in the inventory records.")
    
    inventory = {product: qty for product, qty in inventory.items() if qty > 0}
    
    print("Final filtered inventory:", inventory)
    return inventory

# Test:
inventory = {"t-shirt": 1, "mug": 5}
customer_orders = {"t-shirt"}
update_inventory(customer_orders, inventory) 
````

````python
# 5. Print the total price of the customer order.

products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Initialize Inventory
inventory = {p: int(input(f"Enter the quantity of {p}s available: ")) for p in products}

# Get Customer Orders
num_orders = int(input("Enter the number of customer orders: "))
# We use a set to ensure unique products as per your previous logic
customer_orders = {input("Enter the name of a product that a customer wants to order: ") for _ in range(num_orders)}

# Order Stats
total_ordered = len(customer_orders)
percentage_unique = (total_ordered / len(products)) * 100

print("\nOrder Statistics:")
print(f"Total Products Ordered: {total_ordered}")
print(f"Percentage of Unique Products Ordered: {percentage_unique:.1f}")

# Update Inventory & Filter out zeros
print("\nUpdated Inventory:")
for product in customer_orders:
    if product in inventory:
        inventory[product] -= 1

# Dictionary to remove items that reached 0
inventory = {p: q for p, q in inventory.items() if q > 0}

for p, q in inventory.items():
    print(f"{p}: {q}")

# Calculate Total Price
total_price = sum([float(input(f"Enter the price of {product}: ")) for product in customer_orders])

print(f"Total Price: {total_price}")
````
